```python
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base

#Creating sqlalchemy engine
eng=create_engine("sqlite:///:memory:",echo=True)
```

```python
#Constructing Table metadata for the above Dataframe and creating a Table object which represents the Table
from sqlalchemy import Integer,String,Column

Base = declarative_base()

class Adult(Base):
    __tablename__ ='adults'
    
    Age = Column(Integer,primary_key=True)
    workclass = Column(String)
    fnlwgt = Column(Integer)
    education = Column(String)
    educationnum = Column(Integer)
    maritalstatus = Column(String)
    occupation = Column(String)
    relationship = Column(String)
    race = Column(String)
    sex = Column(String)
    capitalgain = Column(Integer)
    capitalloss = Column(Integer)
    hoursperweek = Column(Integer)
    nativecountry = Column(String)
    label = Column(String)
    
    def __repr__(self):
        return "<Adult(Age='%d',workclass='%s',fnlwgt='%d',education='%s',educationnum='%d',maritalstatus='%s',+\
        occupation='%s',relationship='%s',race='%s',sex='%s',capitalgain='%d',capitalloss='%d',hoursperweek='%d',+\
        nativecountry='%s',label='%s')>" % (
        self.Age,self.workclass,self.fnlwgt,self.education,self.educationnum,self.maritalstatus,self.occupation,\
        self.relationship,self.race,self.sex,self.capitalgain,self.capitalloss,self.hoursperweek,self.nativecountry,self.label)
    
```

```python
#Table Metadata
Adult.__table__
```

```python
#Creating the Adult Table
Base.metadata.create_all(eng)
```

```python
#Creating a new session and binding it to the Engine
from sqlalchemy.orm import sessionmaker
newSession=sessionmaker(bind=eng)
```

```python
newSessionInst=newSession()
```

```python
#Inserting two records into the table
adult1=Adult(Age=32,workclass='Private',fnlwgt=215646,education='HS-grad',educationnum=9,maritalstatus='Divorced',\
        occupation='cleaners',relationship='Not-in-family',race='White',sex='Male',capitalgain=0,capitalloss=0,hoursperweek=40,\
        nativecountry='India',label='<=50K')
adult2=Adult(Age=69,workclass='Private',fnlwgt=815676,education='HS-grad',educationnum=9,maritalstatus='Married',\
        occupation='cleaners',relationship='Not-in-family',race='Black',sex='Female',capitalgain=0,capitalloss=0,hoursperweek=40,\
        nativecountry='India',label='<=50K')

newSessionInst.add_all([adult1,adult2])
```

```python
#newSessionInst.rollback()
newSessionInst.query(Adult).count()
```

```python
#Using filter to update the record
newSessionInst.query(Adult).filter(Adult.Age==32).update({Adult.fnlwgt:333333})
```

```python
#Using Filter to update another record
newSessionInst.query(Adult).filter(Adult.Age==69).update({Adult.maritalstatus:'Divorced'})
```

```python
#retirieving the updated records
newSessionInst.query(Adult).all()
```

```python
#Inserting another two records to test the delete function
adult1=Adult(Age=55,workclass='Private',fnlwgt=215646,education='HS-grad',educationnum=9,maritalstatus='Divorced',\
        occupation='cleaners',relationship='Not-in-family',race='White',sex='Male',capitalgain=0,capitalloss=0,hoursperweek=40,\
        nativecountry='India',label='<=50K')
adult2=Adult(Age=44,workclass='Private',fnlwgt=815676,education='HS-grad',educationnum=9,maritalstatus='Married',\
        occupation='cleaners',relationship='Not-in-family',race='Black',sex='Female',capitalgain=0,capitalloss=0,hoursperweek=40,\
        nativecountry='India',label='<=50K')

newSessionInst.add_all([adult1,adult2])
```

```python
#newSessionInst.rollback()
newSessionInst.query(Adult).all()
```

```python
#Using the execute method to execute the Select query
conn = eng.connect()
result=conn.execute("select * from adults")
result.fetchall()
```

```python
#Deleting the two records inserted earlier
conn.execute("Delete from adults where Age in(44,55)")
```

```python
#retrieving the records after executing delete function
result=conn.execute("select * from adults")
result.fetchall()
```

```python
#Using a function query -func.max
from sqlalchemy import func
newSessionInst.query(func.max(Adult.Age)).all()
```

```python
#Using a function query func.avg
newSessionInst.query(func.avg(Adult.Age)).all()
```
